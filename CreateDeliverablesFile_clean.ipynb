```python
# Define params & constants

# ENTER YOUR ACCESS TOKEN HERE!
access_token = "nnnnnnnnnn" # String
order_by = "due_at" # Sring
pagination = 100 # Int
base_url = "https://canvas.uw.edu/api/v1/courses/"
```

```python
# This was easier. I'm sure there's a way to narrow the course request to the a specific period but I don't know it.
# These are the courses for C5 this quarter. This can be updated to other cohorts or future quarters manually using
# using canvas to fetch names and IDs.
courses_data = {'Course ID':['1552113', '1553044', '1553445'], 'Course Name':['FIN 505', 'OPMGT 520', 'QMETH 505']}
courses_df = pd.DataFrame(courses_data)
```

```python
# Create the master pandas DF

# Create empty dataframe
column_names = ["Due Date", "Course Name", "Assignment Name", "Link"]
assignments_output_df = pd.DataFrame(columns=column_names)

# Iterate over the courses dataframe
for n in range(len(courses_df)):
    print(n)  # Print idx number to confirm all 3 courses.
    course_id = courses_df["Course ID"][n]
    course_name = courses_df["Course Name"][n]
    
    request_url = f'{base_url}{course_id}/assignments?order_by={order_by}&per_page={pagination}&access_token={access_token}'
    df = requests.get(request_url).json() # Get json object from API
    df = pd.json_normalize(df) # Flatten json object
    df = df[["due_at", "name", "html_url"]] # Reduce dataframe to only needed columns
    
    # Iterate over the rows of the temporary dataframe
    for idx, row in df.iterrows():
        new_row = {
            'Due Date':row["due_at"], 
            'Course Name':course_name, 
            'Assignment Name':row["name"], 
            'Link':row["html_url"]}
        
        # Append the output dataframe with the new row.
        assignments_output_df = assignments_output_df.append(new_row, ignore_index=True)
        
```

```python
assignments_output_df['Due Date'] = pd.to_datetime(assignments_output_df["Due Date"]).dt.tz_convert('US/Pacific').dt.tz_localize(None)
```

```python
# Output to CSV
# assignments_output_df.to_csv("master_assignments_list.csv", index=False)

# Output to XLSX
# NOTE: This will possibly require you to install the python package 'openpyxl'
# This can be done with pip 'pip3 install openpyxl'
assignments_output_df.to_excel("master_assignments_list.xlsx", index=False)
```
